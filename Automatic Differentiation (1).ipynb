##### Differentiation of a Exponential function

```julia
#Define a function
function f(x)
    return exp(x)
end
```

```julia
# For Derivative 
function Derf(x)
    return exp(x)
end
```

```julia
#set the point and step size
x = 2.0
h = range(0.001,0.1, length=10)
```

#### Forward Differntiation 

In Forward Differentiation the value converges from the higher value to exact value 
\begin{equation}
    f'(x) \approx \frac{f(x+h) - f(x)}{h},
\end{equation}

```julia
function forward_derivative(f,x,h)
    return (f(x+h)-f(x)) / h
end
ForwardDer = forward_derivative.(f,x,h)
```

```julia
ExactDer = Derf(2)*ones(length(ForwardDer))
```

#### Backward(Reverse)Differentiation
In reverse Differentiation the value converges from the lesser value to exact value
\begin{equation} 
    f'(x) \approx \frac{f(x-h)-f(x)}{h}
    \end{equation}

```julia
function backward_derivative(f,x,h)
    return ((f(x)-f(x-h))) / h
end
    BackwardDer = backward_derivative.(f,x,h)
```

```julia
using Plots
```

```julia
plot(h,[BackwardDer ForwardDer ExactDer] , label=["BackwardDer" "ForwardDer" "Exact"] , lw = [1  1.5 2])
scatter!(h,[BackwardDer ForwardDer ExactDer], label = ["Backward" "Forward" "Exact"])
```

#### Differentiation of Logarthm function

```julia
function y(x)
    return log(x)
end 

function Dery(x)
    return 1/(x)
end
```

```julia
x = 2.0
h = range(0.001,0.1, length=10)
```

```julia
function forward_derivative(y,x,h)
    return (y(x+h)-y(x)) / h
end
ForwardDer = forward_derivative.(y,x,h)
```

```julia
ExactDer = Dery(2)*ones(length(ForwardDer))
```

```julia
function backward_derivative(y,x,h)
    return ((y(x)-y(x-h))) / h
end
    BackwardDer = backward_derivative.(y,x,h)
```

```julia
plot(h,[BackwardDer ForwardDer ExactDer] , label=["BackwardDer" "ForwardDer" "Exact"], lw = [1 1.5 2])

scatter!(h,[BackwardDer ForwardDer ExactDer], label = ["Backward" "Forward" "Exact"])
```
